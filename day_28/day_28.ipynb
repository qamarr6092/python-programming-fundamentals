```python
import numpy as np
array = np.array ( [x for x in range(1,11) ] )
print(array)

#The array() method takes an object and coverts into an array

arr_2 =  np.array ( tuple(x**2 for x in range ( 1 ,11 )) )
print(arr_2)

#The arange method() takes start stop , step argumetns too
arr_3 = np.arange ( 2 , 21 , 2 )
print(arr_3)


#Perfoming vectorized operation
arr_3*= 10
print(arr_3)
```
