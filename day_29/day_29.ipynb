```python
class Dog () :
    def __init__ ( self , name ) :
        self.name = name

    def speak ( self ) :
        print(f'{self.name} is speaking' )

def my_fn ( obj , method ) :
    method(obj)

dog  = Dog('harry')
my_fn( dog , Dog.speak)

def create_and_speak ( cls , name ) :
    ins = cls(name)
    ins.speak()
create_and_speak ( Dog , 'rex' )

#Passing instance or method as refernce in arguments causes us to decide what to do with it later
```
