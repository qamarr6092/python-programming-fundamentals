### Broadcasting the array

Rules of Broadcasting

Broadcasting in NumPy follows a strict set of rules to determine the interaction
between the two arrays:

• Rule 1: If the two arrays differ in their number of dimensions, the shape of the
one with fewer dimensions is padded with ones on its leading (left) side.

• Rule 2: If the shape of the two arrays does not match in any dimension, the array
with shape equal to 1 in that dimension is stretched to match the other shape.

• Rule 3: If in any dimension the sizes disagree and neither is equal to 1, an error is
raised.

This indexing array has been broadcasted 
$\begin{pmatrix} 2 & 2 \\ 0 & 0 \end{pmatrix}$ $\begin{pmatrix} 1 & 0 \\ 1 & 0 \end{pmatrix}$

$\text{Mean} = \begin{pmatrix} 170 & 70 \end{pmatrix}$ height and weight respectively

$\text{cov} = \begin{pmatrix} 100 & 80 \\ 80 & 144 \end{pmatrix}$ main diagonal is variance of height and weight
other diagonal is co variance , like correlation

```python
mean = np.array([170,70])
cov = np.array([[100,80],
                [80,144]
               ])
x = rng.multivariate_normal( mean , cov, size = 100) 
x
                 
```

```python
plt.scatter(x[:,0] , x[:,1] )
plt.ylabel('Weight')
plt.xlabel('Height in cm')
plt.show()
```
