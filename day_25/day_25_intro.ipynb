```python
def is_iterable ( obj ) :
    try :
        iter(obj)
        return True
    except ValueError :
        return False

is_iterable ( [ 1 , 2 , 4 ])
#iterator that works in a for loop
#iterable that produces elements one by one and does not go point backwards it uses next ()
```
