```python
words = [ 'apple' , 'ball' ,  'cat', 'dog' , 'animal' , 'bat' , 'big' ,'cow' , 'crow' ]
by_letter = { }
for word in words :
    letter =  word[0]
    if letter not in by_letter :
        by_letter [ letter ] = [ word ] 
    else :
        by_letter[ letter ] .append(word)
by_letter
```

```python
from itertools import groupby
data = [1, 1, 1, 2  ,1, 2, 3, 1, 1]
for  key , group in groupby ( data ) :
    print( key , list( group))
#so groupby method from the itertools only group the element that are same and next to each other



fav_sing = [ 'udit' , 'kumar' , 'shan' , 'shan' ]
for key , group in groupby ( fav_sing , key = lambda x  : x[0] ):
    print ( key , '-->' , list( group))
#key is the value being grouped
```
